```python
# Importing libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading data
dataset = load_dataset('lukebarousse/data_jobs', split='train')
df = dataset.to_pandas()

# Data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.sample(5)
```

```python
df.index
```

```python
df.index.name = 'job_index' # Set name for index column
```

```python
df.index.dtype
```

```python
df.sample(5)
```

```python
# reset_index(): Resets the DataFrame’s index to the default integer index. 
# This is particularly useful after operations that alter the index, like sorting or filtering, to simplify further data manipulation.
df_usa = df[df['job_country'] == 'United States']
df_usa.index
df_usa.reset_index(inplace=True)
df_usa.index
```

set_index(): Sets one or more existing columns as the index of the DataFrame. 
This is useful for timeseries data or when you want to index by specific attributes.

```python
df_usa.set_index('job_index', inplace=True)
```

sort_index(): Sorts the DataFrame by the index (row labels), either ascending or descending. 
This helps in quickly organizing data by the index and is often used after set_index()
