```python
# Importing libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading data
dataset = load_dataset('lukebarousse/data_jobs', split='train')
df = dataset.to_pandas()

# Data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# pivot_table(values='column_to_aggregate', index='row_index', columns='column_index', aggfunc='mean')
```

```python
median_pivot = df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median').sort_values(by='salary_year_avg', ascending=False)
median_pivot.index.dtype
```

```python
df.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False) # Similar to the code above
```

```python
df.pivot_table(index='job_title_short', aggfunc='size')
```

```python
df.groupby('job_title_short').size() # Similar to the code above
```

```python
df.groupby(['job_country', 'job_title_short'])['salary_year_avg'].agg(['max', 'median', 'min']).dropna()
```

```python
df.pivot_table(
    values='salary_year_avg',
    index=['job_country', 'job_title_short'],
    aggfunc = {'salary_year_avg': ['max', 'median', 'min']}
) # to create the same pivot table like above but the code is much clearer
```

```python
df_job_country_salary = df.pivot_table(
    values='salary_year_avg',
    index='job_country',
    columns='job_title_short',
    aggfunc='median'
)
df_job_country_salary
```

```python
top_countries = df['job_country'].value_counts().head(6).index
df_job_country_salary = df_job_country_salary.loc[top_countries]
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']
df_job_country_salary = df_job_country_salary[job_titles]
df_job_country_salary.plot(kind='bar')
plt.ylabel('Median Salary ($USD)')
plt.xlabel('')
plt.xticks(rotation=45, ha='right')
plt.show()
```
