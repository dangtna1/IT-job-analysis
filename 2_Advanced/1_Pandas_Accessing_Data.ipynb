```python
# Importing libraries
import pandas as pd
from datasets import load_dataset

# Loading data
dataset = load_dataset('lukebarousse/data_jobs', split='train')
df = dataset.to_pandas()

# Data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# df.iloc[rowIndex, columnIndex] where rowIndex, columnIndex could be an index, list, range
df.iloc[0, :] # row 0, all columns
df.iloc[2, 15] # row 3rd, column 16th
df.iloc[2:4, 15:17] 
df.iloc[[2, 3], [15, 16]] # the same as the line above
```

```python
# df.loc similar to df.iloc but can be used with labels instead of just number
df.loc[0]
df.loc[:9, ['job_skills', 'job_type_skills']] # end-value inclusive
df.loc[10:20, 'job_title_short':'job_work_from_home']
```
