```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast

# Loading data
dataset = load_dataset('lukebarousse/data_jobs', split='train')
df = dataset.to_pandas()

# Data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')]
df_DA_US = df_DA_US.dropna(subset='salary_year_avg')
df_DA_US
```

```python
df_DA_US = df_DA_US.explode('job_skills')
df_DA_US[['salary_year_avg', 'job_skills']]
```

```python
df_DA_top_pay = df_DA_US.groupby('job_skills').agg(
    count = ('job_skills','count'),
    median = ('salary_year_avg', 'median')
).sort_values(by='median', ascending=False).head(10)
df_DA_top_pay
```

```python
df_DA_skills = df_DA_US.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='count', ascending=False).head(10)
df_DA_skills
```

```python
type(df_DA_top_pay)
```

```python
fig, ax = plt.subplots(2, 1)

df_DA_top_pay[::-1].plot(kind='barh', y='median', ax=ax[0], legend=False)
ax[0].set_title('Top 10 highest paid skills for data analyst')
ax[0].set_ylabel('')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda value, pos: f'${int(value/1000)}K'))

df_DA_skills.sort_values('median', ascending=False).plot(kind='barh', y='median', ax=ax[1], legend=False)
ax[1].set_title('Top 10 most in-demand skills for data analyst')
ax[1].set_ylabel('')
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda value, pos: f'${int(value/1000)}K'))
ax[1].set_xlim(0, 200000)
ax[1].invert_yaxis()
ax[1].set_xlabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```
